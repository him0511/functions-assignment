1. What is the difference between a function and a method in Python?
- In Python, the terms **function** and **method** refer to two similar but distinct concepts. Here’s the key difference:

### 1. **Function**:
- A **function** is a block of code that performs a specific task. It can be called by its name, and can take input arguments (parameters) and return a value.
- Functions are defined using the `def` keyword, and they can exist independently of any object.
  
Example:
```python
def add(a, b):
    return a + b

result = add(5, 3)  # Calling the function
print(result)  # Output: 8
```
In this example, `add` is a function that is not associated with any object.

### 2. **Method**:
- A **method** is a function that is **associated with an object** and is called on that object. Methods are functions that belong to a class (or an object) and operate on that class's or object's data.
- Methods are typically defined within a class using the `def` keyword, and they must always include `self` as their first parameter to refer to the object instance.

Example:
```python
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
result = calc.add(5, 3)  # Calling the method on an object
print(result)  # Output: 8
```
In this example, `add` is a **method** because it is defined inside a class (`Calculator`) and is called on an instance of that class (`calc`).

### Key Differences:
- **Functions** can be called independently.
- **Methods** are tied to objects and can be called using dot notation on the object (`object.method()`).


2. Explain the concept of function arguments and parameters in Python.
- In Python, **function arguments** and **parameters** refer to the values passed to a function and the variables that receive those values, respectively. Understanding how they work is crucial to writing effective and flexible functions. Here's a detailed explanation:

### 1. **Parameters**:
- **Parameters** are the **variables** listed in the function definition that act as placeholders for the values that will be passed when the function is called.
- They are part of the function's signature and define what inputs the function expects.

Example:
```python
def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")
```
In the above function `greet`, `name` is a parameter.

### 2. **Arguments**:
- **Arguments** are the **actual values** you pass to a function when calling it. These values get assigned to the function's parameters.
- Arguments can be passed either **positionally** or **by keyword**.

#### 2.1. **Positional Arguments**:
- The order in which you pass the arguments matters. The first argument will be assigned to the first parameter, the second to the second parameter, and so on.

Example:
```python
def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")

greet("Alice", 30)  # "Alice" is the first argument (assigned to 'name'), and 30 is the second argument (assigned to 'age')
```
Here, `"Alice"` and `30` are **positional arguments** passed to the parameters `name` and `age` respectively.

#### 2.2. **Keyword Arguments**:
- You can explicitly specify which parameter each argument corresponds to by using the parameter name. This eliminates the need to follow the parameter order.

Example:
```python
def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")

greet(age=30, name="Alice")  # Arguments passed by keyword
```
In this case, the order of arguments doesn't matter because we are using the parameter names (`name` and `age`) to assign values.

### 3. **Default Parameters**:
- You can specify default values for parameters. If the caller doesn't provide a value for such a parameter, the default value is used.

Example:
```python
def greet(name, age=25):  # 'age' has a default value of 25
    print(f"Hello, {name}! You are {age} years old.")

greet("Alice")  # Uses the default value of 25 for 'age'
greet("Bob", 30)  # Overrides the default value with 30
```
In this example, `age` has a default value of 25. If no argument is passed for `age`, the default is used.

### 4. **Variable-Length Arguments**:
Sometimes you don’t know beforehand how many arguments will be passed to your function. Python allows you to handle this with **arbitrary argument lists**.

#### 4.1. **Arbitrary Positional Arguments (`*args`)**:
- Use `*args` to pass a variable number of positional arguments. Inside the function, `args` will be a tuple containing all the passed arguments.

Example:
```python
def greet(*names):
    for name in names:
        print(f"Hello, {name}!")

greet("Alice", "Bob", "Charlie")  # 'names' will be ('Alice', 'Bob', 'Charlie')
```

#### 4.2. **Arbitrary Keyword Arguments (`**kwargs`)**:
- Use `**kwargs` to pass a variable number of keyword arguments. Inside the function, `kwargs` will be a dictionary containing all the passed keyword arguments.

Example:
```python
def greet(**people):
    for name, age in people.items():
        print(f"Hello, {name}! You are {age} years old.")

greet(Alice=30, Bob=25, Charlie=35)
```
In this case, `kwargs` is a dictionary like `{'Alice': 30, 'Bob': 25, 'Charlie': 35}`.

4. What is the purpose of the `return` statement in a Python function?

-  The **`return`** statement in a Python function serves the following primary purposes:

### 1. **Returning a Value from the Function**:
The most common use of the `return` statement is to send a value from the function back to the caller. When a function is called, it may perform some computation or process and then return a result, which can be used in further calculations or assigned to a variable.

**Example:**
```python
def add(a, b):
    return a + b  # Returns the sum of a and b

result = add(3, 5)  # The value returned by the function (8) is stored in the result variable
print(result)  # Output: 8
```
In this example, the `return` statement sends the sum of `a` and `b` back to the caller.

### 2. **Terminating Function Execution Early**:
When the `return` statement is executed, the function immediately ends, and the program continues from the point where the function was called. This means that any code written after the `return` statement will not be executed.

**Example:**
```python
def check_positive(number):
    if number <= 0:
        return "Not positive"  # Exits the function immediately
    return "Positive"

result = check_positive(-5)  # Function returns "Not positive" immediately and stops
print(result)  # Output: "Not positive"
```
In this case, if the condition `number <= 0` is met, the function returns early, and no further code in the function is executed.

### 3. **Returning Multiple Values (Tuples)**:
A function can return multiple values by separating them with commas. These values are automatically packed into a **tuple**, which can then be unpacked into variables when the function is called.

**Example:**
```python
def get_coordinates():
    return 10, 20  # Returns a tuple (10, 20)

x, y = get_coordinates()  # Unpacks the tuple into variables
print(x, y)  # Output: 10 20
```
In this example, the `return` statement sends back two values, and they are unpacked into `x` and `y`.

### 4. **Returning `None` (Implicit Return)**:
If no `return` statement is explicitly provided, or if a `return` statement without an expression is used, the function will return **`None`** by default. This indicates that the function does not explicitly return any value.

**Example:**
```python
def greet():
    print("Hello, World!")

result = greet()  # No return statement, so result will be None
print(result)  # Output: None
```
In this case, the `greet` function prints a message but returns `None` because no value is explicitly returned.

### 5. **Returning from a Function Without Any Result**:
In some cases, a function might need to exit without returning any specific value, for example, to indicate an exit or completion without any data.

**Example:**
```python
def print_message(message):
    if not message:
        return  # Returns None implicitly if message is empty and exits the function
    print(message)

print_message("")  # No output, returns None and exits
```
In this case, the function will return `None` and terminate when an empty string is passed.

### Key Points About the `return` Statement:
1. **Ends the function execution**: When `return` is executed, the function terminates immediately, and the rest of the code inside the function is not executed.
2. **Returns a value**: You can return a single value, multiple values (packed in a tuple), or no value (which results in an implicit return of `None`).
3. **Can be used to pass values back to the caller**: The value returned can be used in further expressions or assigned to variables for future use.

In summary, the `return` statement is essential for:
- Returning values from functions.
- Exiting a function early.
- Allowing functions to produce results that can be used elsewhere in your code.

5. 5. What are iterators in Python and how do they differ from iterables?
- In Python, **iterables** and **iterators** are closely related concepts, but they have distinct characteristics and purposes. Here's a detailed explanation:

### 1. **Iterable**:
An **iterable** is any Python object that can return an iterator. In other words, an iterable is an object that can be looped over (iterated) using a `for` loop or other iteration methods.

- **Key property**: An iterable can be used in a `for` loop or any context where iteration is needed.
- Examples of iterables include lists, tuples, dictionaries, strings, and sets.

#### How an iterable works:
- When an iterable object is passed to the `iter()` function, it returns an **iterator**.
- An iterable does not itself maintain the state of iteration; it just defines how to retrieve the next item.

**Example of an iterable**:
```python
# List is an iterable
my_list = [1, 2, 3]

# The list is iterable, meaning it can be used in a for loop
for item in my_list:
    print(item)
```
In the above example, `my_list` is an iterable.

### 2. **Iterator**:
An **iterator** is an object that represents a stream of data, and it maintains the state of iteration. An iterator knows how to get the next item in a sequence and knows when the sequence has been exhausted.

- **Key property**: An iterator is an object that performs the actual iteration over the iterable, keeping track of the current position in the iteration.
- An iterator implements two methods:
  - `__iter__()`: This method returns the iterator object itself. It's required to make an object an iterator.
  - `__next__()`: This method returns the next item from the iterable. When there are no more items to return, it raises a `StopIteration` exception.

#### How an iterator works:
- Once you have an iterator (via the `iter()` function on an iterable), you can retrieve items using the `next()` function or a `for` loop.
- An iterator can only be traversed once, and after it is exhausted, it cannot be reused or reset.

**Example of an iterator**:
```python
# List is an iterable
my_list = [1, 2, 3]

# Get an iterator from the iterable using iter()
my_iterator = iter(my_list)

# Use next() to iterate through the items
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3

# Trying to get the next item after the iterable is exhausted will raise StopIteration
# print(next(my_iterator))  # Raises StopIteration
```
In the example above, `my_iterator` is an iterator, and `next()` is used to manually retrieve each item from the iterable.

### 3. **Key Differences Between Iterables and Iterators**:

| **Feature**         | **Iterable**                                    | **Iterator**                              |
|---------------------|-------------------------------------------------|-------------------------------------------|
| **Definition**      | An object that can return an iterator using `iter()` | An object that tracks the iteration state |
| **Methods**         | Implements `__iter__()` method, but not `__next__()` | Implements both `__iter__()` and `__next__()` |
| **Purpose**         | Provides a sequence of items that can be iterated over | Provides the mechanism to iterate through the sequence |
| **State**           | Does not maintain iteration state (just provides data) | Maintains the state of the current position in the iteration |
| **Usage**           | Can be passed to a `for` loop or to the `iter()` function | Requires `next()` to fetch items and raises `StopIteration` when exhausted |
| **Examples**        | Lists, tuples, dictionaries, sets, strings       | An iterator object returned from `iter()` or created manually |

### 4. **How to Create Your Own Iterator**:
You can create your own custom iterator by defining a class with the `__iter__()` and `__next__()` methods.

**Example of a custom iterator**:
```python
class Countdown:
    def __init__(self, start):
        self.start = start
    
    def __iter__(self):
        self.current = self.start  # Initialize the starting point
        return self
    
    def __next__(self):
        if self.current <= 0:
            raise StopIteration  # End the iteration when countdown reaches 0
        self.current -= 1
        return self.current

# Using the Countdown iterator
countdown = Countdown(5)
for num in countdown:
    print(num)
```
In this example, the `Countdown` class is an iterable that counts down from a specified number. It creates an iterator when used in a `for` loop.

### Summary:
- **Iterable**: An object that can return an iterator (e.g., lists, strings). It can be passed to the `iter()` function, which gives an iterator to traverse the object.
- **Iterator**: An object that actually performs the iteration and keeps track of the current position in the iteration. It implements `__iter__()` and `__next__()`.

In short:
- **Iterables** provide the sequence to be traversed.
- **Iterators** perform the traversal and track the iteration state.

6. Explain the concept of generators in Python and how they are defined.
- In Python, a **generator** is a special type of iterator that allows you to create iterators in a more concise and efficient way. Instead of returning all items at once, generators **lazily produce** items one at a time only when required. This makes them more memory-efficient, especially when working with large datasets or infinite sequences.

### 1. **What Are Generators?**
A **generator** is a function that uses the `yield` keyword to return values one at a time. Each time the `yield` statement is executed, the function’s state is paused, and the value is returned to the caller. The state of the function (including local variables and the position of execution) is preserved, so when the generator is called again, it picks up where it left off.

This "lazy evaluation" allows generators to produce items on demand and is ideal for working with large data sets or infinite sequences, as they do not require all items to be stored in memory at once.

### 2. **How to Define a Generator?**
A generator is defined just like a normal function, except that it uses the `yield` statement instead of `return` to yield values one at a time.

**Syntax:**
```python
def generator_function():
    # Some code
    yield value
    # Some more code
    yield another_value
```

### 3. **Example of a Generator:**

```python
def countdown(start):
    while start > 0:
        yield start
        start -= 1

# Create a generator object
gen = countdown(5)

# Using the generator in a for loop
for num in gen:
    print(num)
```

**Output:**
```
5
4
3
2
1
```

In this example:
- The `countdown` function is a generator because it uses `yield` to return values one by one.
- When the `for` loop iterates over the generator, the `countdown` function is called repeatedly and yields the next value in the sequence.

### 4. **How Generators Work**:
- **`yield` vs `return`**: The `yield` keyword returns a value from the generator, but instead of terminating the function, it pauses the function’s execution, saving its state. The function can resume from where it left off when the generator is called again.
- When the generator is exhausted (i.e., there are no more values to yield), it raises a `StopIteration` exception, signaling the end of iteration.

### 5. **Characteristics of Generators**:
- **Lazy evaluation**: Generators only compute the next value when required, so they are memory-efficient, especially with large datasets.
- **Stateful**: Generators maintain their state between calls, so they can pick up where they left off.
- **Exhaustible**: Once a generator is exhausted (i.e., all values have been yielded), it cannot be reused. If you want to iterate over the values again, you need to create a new generator.

### 6. **Using the `next()` Function with Generators**:
You can manually retrieve the next value from a generator using the `next()` function. If the generator is exhausted, `StopIteration` is raised.

**Example:**
```python
def simple_generator():
    yield 1
    yield 2
    yield 3

gen = simple_generator()

# Using next() to manually get values
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3

# After the generator is exhausted, it raises StopIteration
# print(next(gen))  # Raises StopIteration
```

### 7. **Generator Expressions**:
Python also provides a concise way to create generators using **generator expressions**, which are similar to list comprehensions but use parentheses `()` instead of square brackets `[]`.

**Example:**
```python
# Generator expression
gen = (x * x for x in range(5))

# Using the generator in a for loop
for num in gen:
    print(num)
```

**Output:**
```
0
1
4
9
16
```

Here, the generator expression `(x * x for x in range(5))` creates a generator that yields the squares of numbers from 0 to 4.

### 8. **Advantages of Generators**:
- **Memory Efficiency**: Since generators produce one value at a time and do not store all items in memory, they are more memory-efficient, especially when working with large datasets.
- **Lazy Evaluation**: They compute values on demand, which means you can start processing items immediately without waiting for the entire collection to be created.
- **Infinite Sequences**: Generators are ideal for creating infinite sequences because they can keep yielding values indefinitely without consuming infinite memory.

### 9. **Use Cases for Generators**:
- **Processing large datasets**: For example, reading large files line by line or generating a sequence of numbers without loading the entire sequence into memory.
- **Streaming data**: Useful when working with data that is produced or consumed in streams, such as logs, real-time data feeds, etc.
- **Pipelines**: In scenarios where data needs to be processed through a series of transformations, generators can be used to chain processing steps efficiently.

### Example of a generator used for reading large files:
```python
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Yield each line, one at a time

# Use the generator to process large files without loading everything into memory
for line in read_large_file('large_file.txt'):
    print(line)
```

### Summary:
- A **generator** is a type of iterator that uses the `yield` keyword to produce values lazily, one at a time.
- **Generators** are more memory-efficient because they compute values on demand and don't store the entire sequence in memory.
- Generators are defined like regular functions but use `yield` instead of `return`.
- **Generator expressions** are a concise way to create generators.


7. What are the advantages of using generators over regular functions?
- Generators offer several key **advantages** over regular functions, particularly when dealing with large datasets, infinite sequences, or situations where you want to **optimize memory usage**. Here's a detailed breakdown of the advantages of using generators:

### 1. **Memory Efficiency**:
- **Generators produce values lazily**, meaning they generate each item one at a time only when required (on-demand). This is particularly useful when dealing with large datasets or sequences that don't fit into memory all at once.
- **Regular functions**, on the other hand, return a complete result all at once (e.g., a list or a tuple), which can consume a significant amount of memory, especially when dealing with large collections.
  
**Example**:
Imagine you're reading a very large file line by line:
- **Using a regular function**: You would need to read all lines into a list, which would consume a lot of memory if the file is large.
- **Using a generator**: The generator reads each line one at a time, keeping only one line in memory at a time, which is much more memory-efficient.

**Memory Efficiency Example with Generator**:
```python
# Generator function for reading large files
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

# Generator processes the file line by line, conserving memory
for line in read_large_file('large_file.txt'):
    print(line)  # Only one line is kept in memory at any time
```

### 2. **Lazy Evaluation**:
- **Generators** allow you to **compute values only when needed**. This is beneficial when you don’t need to calculate or store all values in memory upfront.
- **Regular functions** that return large lists or collections will calculate all values in advance, which may be unnecessary if you're only interested in a small subset of the results at any given time.

**Lazy Evaluation Example**:
```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

gen = countdown(5)

# Lazy evaluation: Values are generated on demand, one by one
print(next(gen))  # Output: 5
print(next(gen))  # Output: 4
```
In this example, the `countdown` generator doesn't calculate all values at once. Instead, it only computes the next number when requested, allowing for **on-demand computation**.

### 3. **Improved Performance with Large Data**:
- **Generators** can be much faster in scenarios where you only need to process part of a dataset or stream of data.
- When using regular functions to handle large amounts of data (such as lists), Python has to **create the entire list**, which can be slow and inefficient, especially for very large datasets.
- **Generators avoid this issue** by producing and yielding values one at a time.

**Performance Example**:
Consider processing a large range of numbers, where you only need to perform operations on some of them. A generator will start producing numbers immediately without waiting for the entire collection to be built.

```python
# Generator: yields numbers one at a time
gen = (x * x for x in range(1000000))

# Only the square of the next number is calculated when needed
for _ in range(5):
    print(next(gen))  # Output: 0, 1, 4, 9, 16 (squares of numbers)
```

### 4. **Infinite Sequences**:
- **Generators can represent infinite sequences** because they don’t need to generate or store all the values at once. This is particularly useful when dealing with **infinite data** (like an infinite range of numbers).
- **Regular functions** that return a full list or collection cannot represent infinite sequences, as the entire sequence would need to be stored in memory.

**Infinite Sequence Example with Generator**:
```python
def infinite_numbers():
    n = 0
    while True:
        yield n
        n += 1

gen = infinite_numbers()

# Prints the first 5 numbers from the infinite sequence
for _ in range(5):
    print(next(gen))  # Output: 0, 1, 2, 3, 4
```
Here, the `infinite_numbers` generator produces an infinite sequence of numbers without ever needing to store them all in memory.

### 5. **Clean and Concise Code**:
- **Generators** often provide a cleaner and more concise way to write code that needs to yield values.
- **Regular functions** may require creating complex data structures (like lists) to collect values and then return them all at once, leading to bulkier code.
- With **generators**, you can write a function that yields values one at a time in a more readable and straightforward way, without needing to manage intermediate data structures.

**Concise Code Example with Generator**:
```python
# Generator expression: concise way to create a generator
gen = (x * x for x in range(10))

# Generator processes values lazily in a for loop
for value in gen:
    print(value)  # Output: 0, 1, 4, 9, 16, ...
```

### 6. **Maintaining State**:
- **Generators** maintain their **own state** between successive calls, which allows them to yield values one at a time without recalculating everything. The `yield` statement pauses the generator's execution, storing the state of the function, so when the generator is resumed, it picks up where it left off.
- **Regular functions** don't maintain state unless explicitly stored, and they typically return values all at once, which may require recalculating the results if you need to revisit them.

**State Example with Generator**:
```python
def count_up_to(limit):
    count = 0
    while count < limit:
        yield count
        count += 1

gen = count_up_to(3)

print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
```
The generator `count_up_to` maintains the state of the `count` variable and continues from where it left off when `next()` is called.

### 7. **Reduced Overhead for Large Loops**:

9. Explain the purpose and usage of the `map()` function in Python.
- The `map()` function in Python is a built-in higher-order function that allows you to apply a specified function to each item in one or more iterables (such as a list, tuple, etc.) and return a map object (an iterator) that yields the results. This is particularly useful when you need to perform the same operation on each element in a collection, avoiding the need for an explicit loop.

### 1. **Syntax of `map()`**:

```python
map(function, iterable, ...)
```

- **`function`**: A function that will be applied to each element of the iterable(s). This function should take as many arguments as there are iterables passed to `map()`.
- **`iterable`**: One or more iterables (like lists, tuples, etc.) whose elements will be passed to the `function`.
- **Return value**: The `map()` function returns a **map object** which is an iterator that produces the transformed items when iterated over. It does **not return a list** directly, but it can be converted into a list or other iterable types if needed.

### 2. **Basic Example**:

Let’s start with a simple example to demonstrate how `map()` works:

#### Example 1: Applying a function to each item in a list
```python
# Function to double a number
def double(x):
    return x * 2

# List of numbers
numbers = [1, 2, 3, 4]

# Using map to apply the function 'double' to each element in the list
result = map(double, numbers)

# Convert the result to a list to see the output
print(list(result))  # Output: [2, 4, 6, 8]
```

Here:
- The function `double(x)` is applied to each element of the list `numbers`.
- `map()` returns a **map object** that is then converted into a list to display the results.

### 3. **Using Lambda Functions with `map()`**:

You can often use a **lambda function** (an anonymous, one-line function) directly inside `map()`, which can make your code more concise and readable.

#### Example 2: Using `lambda` to double each number

Here:
- The lambda function `lambda x: x * 2` is used directly inside `map()` to double each element in the `numbers` list.
- As with the previous example, the result is a map object that is converted to a list for easier viewing.

### 4. **Using Multiple Iterables**:

`map()` can take more than one iterable as input. When you pass multiple iterables, the function you provide must accept as many arguments as there are iterables. `map()` will apply the function to elements from the iterables in parallel, processing them element by element.

#### Example 3: Using `map()` with multiple iterables
```python

    return x + y

# Two lists of numbers
list1 = [1, 2, 3, 4]
list2 = [10, 20, 30, 40]

```

In this example:
- `map()` takes two iterables, `list1` and `list2`, and applies the `add()` function to corresponding elements from each list.
- The result is a map object, which is then converted into a list.

### 5. **Working with Other Data Types**:

`map()` can also be used with other data types, not just lists. For example, you can use `map()` to apply a function to the elements of a tuple or a string.

### 6. **Efficiency and Lazy Evaluation**:

One of the key benefits of using `map()` is that it returns an **iterator**, which means that the function is applied lazily—values are computed only when they are needed. This is more memory efficient compared to other methods that generate a full list upfront.

For example, with large datasets or when you don't need all the results at once, `map()` can help reduce memory usage since it doesn't create the entire result set at once.

Here:
- `map()` lazily evaluates the squared numbers, so only the necessary values are calculated when you convert the iterator to a list.

### 7. **When to Use `map()`**:

You should use `map()` when you need to **apply a function to each item in an iterable** and want to avoid writing explicit loops. It's a concise and functional way to transform data. Some common use cases include:
- **Transforming a list or iterable**: When you need to apply a transformation (like doubling, squaring, or changing the case of strings) to each element.
- **Working with multiple iterables**: When you need to combine elements from multiple iterables element-by-element (like adding corresponding elements from two lists).
- **Efficiency**: When dealing with large datasets or when you want to process data lazily (on-demand).

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
The `map()`, `reduce()`, and `filter()` functions in Python are all **higher-order functions** that allow you to process iterables in functional programming style. They take a function and one or more iterables as input, but they each serve different purposes and work in different ways.

### 1. **`map()` Function**:
- **Purpose**: The `map()` function is used to apply a given function to **every item** in an iterable (or multiple iterables) and return an iterator that produces the results.
- **Result**: It returns a **map object** (an iterator), which can be converted into a list or another iterable if needed.

#### Syntax:
```python
map(function, iterable, ...)
```

- **`function`**: The function that will be applied to each element of the iterable.
- **`iterable`**: One or more iterables to process.

#### Example:
```python
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)
print(list(result))  # Output: [2, 4, 6, 8]
```

- **Usage**: Use `map()` when you want to **transform** each element of an iterable using a function.
- **Behavior**: It applies the function **independently** to each element of the iterable.

---

### 2. **`reduce()` Function**:
- **Purpose**: The `reduce()` function is used to **reduce** an iterable to a **single accumulated result** by repeatedly applying a binary function (a function that takes two arguments) to the items of the iterable.
- **Result**: It returns a **single value**, which is the result of repeatedly applying the function to the elements of the iterable.

#### Syntax:
```python
from functools import reduce
reduce(function, iterable, [initial])
```

- **`function`**: A binary function that takes two arguments and returns a result (e.g., `x + y`, `x * y`).
- **`iterable`**: The iterable whose elements are reduced.
- **`initial`**: An optional initial value that is used as the starting point (default is `None`).

#### Example:
```python
from functools import reduce

numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)  # Output: 10 (1 + 2 + 3 + 4)
```

- **Usage**: Use `reduce()` when you want to **combine** the elements of an iterable into a **single result** (e.g., summing, multiplying).
- **Behavior**: It applies the function cumulatively, starting from the first element and reducing them step-by-step.

---

### 3. **`filter()` Function**:
- **Purpose**: The `filter()` function is used to **filter out elements** from an iterable based on a given condition. The condition is provided by a function that returns `True` or `False` for each element.
- **Result**: It returns a **filter object** (an iterator), which can be converted into a list or another iterable if needed. Only elements that satisfy the condition (i.e., when the function returns `True`) are included.

#### Syntax:
```python
filter(function, iterable)
```

- **`function`**: A function that returns a Boolean value (`True` or `False`).
- **`iterable`**: The iterable to process.

#### Example:
```python
numbers = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))  # Output: [2, 4]
```

- **Usage**: Use `filter()` when you want to **select** elements from an iterable based on a condition.
- **Behavior**: It **filters** the iterable by applying the function to each element and keeping only the elements that return `True`.

---

### 4. **Comparison**:

| **Function**  | **Purpose**                                                                 | **Result**                                   | **Example Use Case**                      |
|---------------|-----------------------------------------------------------------------------|---------------------------------------------|-------------------------------------------|
| **`map()`**   | Applies a function to **each element** of an iterable (or multiple iterables). | Transformed iterable                        | Transforming data (e.g., doubling numbers)|
| **`reduce()`**| Applies a binary function to **reduce** an iterable to a **single result**.   | A single accumulated value                  | Summing, multiplying, or combining items |
| **`filter()`**| Filters elements based on a condition (returns `True` or `False`).           | A filtered iterable                         | Selecting elements that meet a condition (e.g., even numbers) |

### 5. **Use Case Scenarios**:


### 6. **Behavior Summary**:
- **`map()`**: Applies a function to each item, returning a transformed iterable.

- **`filter()`**: Filters an iterable based on a condition, returning only the elements that meet the condition.

### 7. **When to Use Each Function**:
- **Use `map()`** when you want to **transform** each element in a collection.
- **Use `reduce()`** when you want to **combine** all elements in an iterable into a single result.
- **Use `filter()`** when you want to **select** certain elements from an iterable based on a condition.

In conclusion, these functions each serve distinct purposes and can be used together or separately depending on the task at hand.

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

````python
class FibonacciIterator:
    def __init__(self, terms):
        # Initialize the number of terms for the Fibonacci sequence
        self.terms = terms
        self.a, self.b = 0, 1  # Starting values of the Fibonacci sequence
        self.count = 0  # Counter to track how many terms have been generated

    def __iter__(self):
        # The __iter__ method should return the iterator object itself
        return self

    def __next__(self):
        # If the count reaches the specified number of terms, stop the iteration
        if self.count >= self.terms:
            raise StopIteration
        # Generate the next Fibonacci number
        fib_number = self.a
        self.a, self.b = self.b, self.a + self.b  # Update Fibonacci sequence
        self.count += 1
        return fib_number
````

````python
# Create an instance of the FibonacciIterator to generate 10 terms
fib_iterator = FibonacciIterator(10)

# Using the iterator to print Fibonacci numbers
for number in fib_iterator:
    print(number)
````
